# String Methods
## Accessing

```python
string='Cloud number 9'
string[0]
```

```python
print(string[0])
```

```python
print(len(string))
```

```python
print(string.find('n'))
```

```python
print(string.count('u'))
```

## Slicing

```python
print(string[0:5])
print(string[:5])
```

```python
print(string[6:9])
```

```python
print(string[6:])
```

```python
print(string[:-6])
```

```python
print(string[-5:])
```

## Transforming

```python
print(string.split(' '))
```

```python
print(string.replace('9','7'))
```

```python
print(string.upper())
print(string.lower())
```

```python
print('-'.join(string))
```

## Checking/Testing

```python
print(string.startswith('Cloud'))
print(string.endswith('o'))
```

```python
print(string.isalnum())  
```

```python
print(string.isdigit())
```

```python
print(string.isspace())
```
