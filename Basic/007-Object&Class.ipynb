```python
class Car:
    def drive(self):
        print(self.brand, 'car is driving')
    def get_speed(self):
        print(self.speed, 'is my top-speed')
    def get_gear(self):
        print(self.gear, 'gear')
        
toyota = Car()
toyota.brand = 'Toyota'
toyota.speed = 150
toyota.gear = 'Auto'

toyota.drive()
toyota.get_speed()
toyota.get_gear()

bmw = Car()
bmw.brand = 'BMW'
bmw.speed = 100
bmw.gear = 'Manual'

bmw.drive()
bmw.get_speed()
bmw.get_gear()
```

## Constructor

```python
class Car:
    def __init__(self, brand, color, speed, gear):
        self.brand = brand
        self.color = color
        self.speed = speed
        self.gear = gear  
        
    def drive(self):
        print(self.brand, 'car is driving')
    def get_speed(self):
        print(self.speed, 'is my top-speed')
    def get_gear(self):
        print(self.gear, 'gear')

    def make_car(self):
        print('I made a new car,', self.brand, self.speed, self.gear)

toyota = Car('Toyota', 'Red', '140', 'Auto')
toyota.make_car()
toyota.drive()
toyota.get_speed()
toyota.get_gear()

bmw = Car('BMW', 'Blue', '100', 'Manual')
bmw.make_car()
bmw.drive()
bmw.get_speed()
bmw.get_gear()
```
