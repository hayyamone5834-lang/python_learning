```python
file_object = open('myFile.txt', 'w+')
L = ['Yangon \n', 'Mandalay \n', 'Pyay \n']
file_object.write('City \n')
file_object.writelines(L)
file_object.close()
```

```python
file_object = open('myFile.txt', 'w+')
L = ['Yangon \n', 'Mandalay \n', 'Pyay \n']
file_object.write('City \n')
file_object.writelines(L)
file_object.seek(0)
print(file_object.read())
file_object.close()
```
