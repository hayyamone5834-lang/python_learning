```python
header = ['Name', 'Department', 'Salary']
data = [['Ag Ag', 'Finance', 350000],['Ni Ni', 'HR', 300000]]
with open('nextTestFile.csv', mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(header)
    csv_writer.writerow(['Win Win', 'Engineering', 500000])
    csv_writer.writerows(data)
```

```python
add_data = [['Kyaw Kyaw', 'Finance', 350000],['Bo Bo', 'HR', 300000]]
with open('nextTestFile.csv', mode='a', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerows(add_data)
```

```python
heading = ['Name', 'Department', 'Salary']
with open('dict_format.csv', mode='w', newline='') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames = heading)
    csv_writer.writeheader()
    csv_writer.writerow({'Name': 'Kyi Kyi', 'Department': 'Accounting', 'Salary': 300000})
    csv_writer.writerow({'Name': 'Myint Myint', 'Department': 'IT', 'Salary': 400000})
```

```python
import csv
rows = []
with open('dict_format.csv') as csv_file:
    csv_reader = csv.reader(csv_file)
    fields = next(csv_reader)
    for row in csv_reader:
        rows.append(row)
print(csv_reader)
print(fields)
print(rows)
```
