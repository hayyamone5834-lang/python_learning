### Replace Null with average values 

```python
data = {
    'Product': ['Milk', 'Coffee', 'Milo', 'Tea'],
    'Price': [500, None, 400, 350]
}
tdf = pd.DataFrame(data)
tdf
```

```python
df_price = tdf['Price']
df_price.describe()
```

```python
df_price.mean()
```

```python
avg_price = df_price.mean()
df_price.fillna(avg_price, inplace=True)
tdf
```
