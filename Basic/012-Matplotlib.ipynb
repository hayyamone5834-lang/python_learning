## Bar Charts

```python
x = ['Samsung','Nokia','Apple','Oppo','Huawei','Vivo']
y = [22,12,14,6,13,18]

plt.bar(x,y)
plt.show()
```

```python
plt.xlabel('Mobile Phone')
plt.ylabel('Usage')
plt.title('Usage of Mobile Phone')
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
phone = ['Samsung','Nokia','Apple','Oppo','Huawei','Vivo']
man = [10,5,4,2,10,10]
woman = [12,7,10,4,3,8]

fig,ax = plt.subplots()

x = np.arange(6)
width = 0.4

ax.bar(x-width/2, man, width, label='Man')
ax.bar(x+width/2, woman, width, label='Woman')
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(phone)
plt.show()
```

```python
phone = ['Samsung','Nokia','Apple','Oppo','Huawei','Vivo']
man = [10,5,4,2,10,10]
woman = [12,7,10,4,3,8]

fig,ax = plt.subplots()

y = np.arange(6)
width = 0.4

ax.barh(y-width/2, man, width, label='Man')
ax.barh(y+width/2, woman, width, label='Woman')
ax.legend()
ax.set_yticks(x)
ax.set_yticklabels(phone)
plt.show()
```

```python
plt.bar(phone, man, color='red')
plt.bar(phone, woman, color='yellow', bottom=man)

plt.xlabel('Brand')
plt.ylabel('Usage')
plt.show()
```

```python
plt.barh(phone, man, color='red')
plt.barh(phone, woman, color='yellow', left=man)

plt.xlabel('Brand')
plt.ylabel('Usage')
plt.show()
```

## Scatter Plot

```python
x = [2,4,6,8,10]
y = [1,3,5,7,9]

plt.scatter(x,y)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randint(1,100,50)
y = rng.randint(1,100,50)

plt.scatter(x,y)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randint(1,100,50)
y = rng.randint(1,100,50)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score - Maths vs Physics')

plt.scatter(x,y)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randint(1,100,50)
y = rng.randint(1,100,50)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score - Maths vs Physics')

plt.scatter(x,y, s=100, color='green', marker='v')
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randint(1,100,50)
y = rng.randint(1,100,50)

sizes = 10 * np.add(x,y)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score - Maths vs Physics')

plt.scatter(x,y, s=sizes, alpha=0.5)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randint(1,100,50)
y = rng.randint(1,100,50)

colors = rng.randint(1,100,50)
sizes = 10 * np.add(x,y)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score - Maths vs Physics')

plt.scatter(x,y, s=sizes, c=colors, alpha=0.5)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randint(1,100,50)
y = rng.randint(1,100,50)

colors = rng.randint(1,100,50)
sizes = 10 * np.add(x,y)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score - Maths vs Physics')

plt.scatter(x,y, s=sizes, c=colors, alpha=0.5, cmap='plasma')
plt.colorbar()
plt.show()
```

## Plotting Histogram

```python
x = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
num_bins = 5
plt.hist(x,num_bins)
plt.show()
```

```python
num_bins = 7
plt.hist(x,num_bins)
plt.show()
```

```python
plt.hist(x, num_bins, color='red', alpha=0.5)
plt.show()
```

```python
plt.hist(x, num_bins, color='red', alpha=0.5, histtype='step')
plt.show()
```

```python
plt.hist(x, num_bins, color='red', alpha=0.5, histtype='bar', orientation='horizontal')
plt.show()
```

```python
x1 = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
x2 = [20,20,30,30,45,45,50,50,65,60,65,75,75,80,80,90,90,35,85]

num_bins = 7
plt.hist([x1,x2], label=['x1','x2'])
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(x1, 5, label='x1', alpha=0.5, color='blue')
plt.hist(x2, 10, label='x2', alpha=0.5, color='red')
plt.legend(loc='upper right')
plt.show()
```

```python
num_bins = 7
plt.hist(x1, num_bins, label='Maths', alpha=0.5, color='blue')
plt.hist(x2, num_bins, label='Physics', alpha=0.5, color='red')
plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student Scores')
plt.legend(loc='upper right')
plt.show()
```

## Plotting Pie Charts

```python
data = [25,75]
plt.pie(data)
plt.show()
```

```python
data = [25,75]
plt.pie(data)
plt.axis('equal')
plt.show()
```

```python
data = [100, 50, 150]
plt.pie(data)
plt.axis('equal')
plt.show()
```

```python
data = [100, 50, 150]
label = ['Coffee', 'Tea', 'Milk']
plt.pie(data, labels=label)
plt.title('Drink Distribution')
plt.axis('equal')
plt.show()
```

```python
data = [100, 50, 150]
label = ['Coffee', 'Tea', 'Milk']
color = ['brown', 'orange', 'gray']
plt.pie(data, labels=label, colors=color)
plt.title('Drink Distribution')
plt.axis('equal')
plt.show()
```

```python
data = [100, 50, 150]
label = ['Coffee', 'Tea', 'Milk']
color = ['brown', 'orange', 'gray']
separate = [0,0.1,0]
plt.pie(data, labels=label, colors=color, explode=separate)
plt.title('Drink Distribution')
plt.axis('equal')
plt.show()
```

```python
data = [100, 50, 150]
label = ['Coffee', 'Tea', 'Milk']
color = ['brown', 'orange', 'gray']
separate = [0,0.1,0]
plt.pie(data, labels=label, colors=color, explode=separate, shadow=True)
plt.title('Drink Distribution')
plt.axis('equal')
plt.show()
```

```python
data = [100, 50, 150]
label = ['Coffee', 'Tea', 'Milk']
color = ['brown', 'orange', 'gray']
separate = [0,0.1,0]
plt.pie(data, labels=label, colors=color, explode=separate, shadow=True, autopct='%1.2f%%')
plt.title('Drink Distribution')
plt.axis('equal')
plt.show()
```

## Plotting Tables

```python
data = [
    [1,2,3,4,5],
    [2,4,6,8,10],
    [1,3,5,7,9]
]
fig, ax = plt.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)
table = ax.table(cellText=data)
table.scale(1,10)
plt.show()
```

```python
data = [
    [1,2,3,4,5],
    [2,4,6,8,10],
    [1,3,5,7,9]
]
cols = ['One', 'Two', 'Three', 'Four', 'Five']
fig, ax = plt.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)
table = ax.table(cellText=data, colLabels=cols, loc='upper center', cellLoc='center')
table.scale(1,10)
plt.show()
```

```python
data = [
    [1,2,3,4,5],
    [2,4,6,8,10],
    [1,3,5,7,9]
]
cols = ['One', 'Two', 'Three', 'Four', 'Five']
rows = ['RowOne', 'RowTwo', 'RowThree']
fig, ax = plt.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)
table = ax.table(cellText=data, rowLabels=rows, colLabels=cols, loc='upper center', cellLoc='center')
table.scale(1,20)
plt.show()
```

```python
data = [
    [1,2,3,4,5],
    [2,4,6,8,10],
    [1,3,5,7,9]
]
cols = ['One', 'Two', 'Three', 'Four', 'Five']
rows = ['RowOne', 'RowTwo', 'RowThree']
Ccolor = ['Orange', 'Orange', 'Orange', 'Orange', 'Orange']
Rcolor = ['cyan', 'cyan', 'cyan']
fig, ax = plt.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)
table = ax.table(cellText=data, rowLabels=rows, colLabels=cols, loc='upper center', cellLoc='center', colColours=Ccolor, rowColours=Rcolor)
table.scale(1,20)
plt.show()
```

```python
data = [
    [1,2,3,4,5],
    [2,4,6,8,10],
    [1,3,5,7,9]
]
cols = ['One', 'Two', 'Three', 'Four', 'Five']
rows = ['RowOne', 'RowTwo', 'RowThree']
Ccolor = ['Orange', 'Orange', 'Orange', 'Orange', 'Orange']
Rcolor = ['cyan', 'cyan', 'cyan']

colors = [
    ['red','w','w','w','w'],
    ['w','w','green','w','w'],
    ['w','w','w','w','blue']
]

fig, ax = plt.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)
table = ax.table(cellText=data, rowLabels=rows, colLabels=cols, loc='upper center', cellLoc='center', colColours=Ccolor, rowColours=Rcolor, cellColours=colors)
table.scale(1,20)
plt.show()
```

```python
data = [
    [1,2,3,4,5],
    [2,4,6,8,10],
    [1,3,5,7,9]
]
cols = ['One', 'Two', 'Three', 'Four', 'Five']
rows = ['RowOne', 'RowTwo', 'RowThree']
Ccolor = ['Orange', 'Orange', 'Orange', 'Orange', 'Orange']
Rcolor = ['cyan', 'cyan', 'cyan']

colors = [
    ['red','w','w','w','w'],
    ['w','w','green','w','w'],
    ['w','w','w','w','blue']
]

fig, ax = plt.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)
table = ax.table(cellText=data, rowLabels=rows, colLabels=cols, loc='upper center', cellLoc='center', colColours=Ccolor, rowColours=Rcolor, cellColours=colors)
table.scale(1,20)
plt.savefig('table.png', format='png')
plt.show()
```
