```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://gad.gov.mm/district-townships.php?stateID=3')
soup = BeautifulSoup(page.content, 'html.parser')
print(soup)
```

```python
print(soup.find('table'))
```

```python
table = soup.find('table')
trs = table.find_all('tr')

for tr in trs:
    for td in tr.find_all('td'):
        print(td)
```

```python
table = soup.find('table')
trs = table.find_all('tr')

rows = []

for tr in trs:
    for td in tr.find_all('td'):
        rows.append(td.get_text(strip=True))
rows
```

```python
table = soup.find('table')
trs = table.find_all('tr')

rows = []

for tr in trs:
    for td in tr.find_all('td'):
        rows.append([td.get_text(strip=True)])
rows
```

```python
table = soup.find('table')
trs = table.find_all('tr')

rows = []

for tr in trs:
    rows.append([td.get_text(strip=True) for td in tr.find_all('td')])
rows
```

```python
import pandas as pd
dfTable = pd.DataFrame(rows[1:], columns=rows[1])
dfTable.head()
```

```python
dfTable.to_csv('gad.csv')
```
