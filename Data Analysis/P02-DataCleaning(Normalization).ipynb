```python
import pandas as pd

data = pd.read_csv('sample-clean.csv')
data.head()
```

```python
data['height'].max()
```

```python
data['height'].min()
```

### Simple Featured Scaling

```python
data['height'] = data['height']/data['height'].max()
data.head()
```

### Min-Max Normalization

```python
data['weight'] = (data['weight'] - data['weight'].min())/(data['weight'].max() - data['weight'].min())
data.head()
```

```python
data.describe()
```

### Z-score Normalization

```python
data['salary'] = (data['salary'] - data['salary'].mean())/data['salary'].std()
data.head()
```

### Convert Category to Numeric

```python
pd.get_dummies(data, columns=['gender'])
```

```python
data.head()
```

### Binning

```python
import pandas as pd

data = pd.read_csv('sample-clean.csv')
data.head()
```

```python
import numpy as np

bins = np.linspace(min(data['salary']), max(data['salary']), 6)
bins
```

```python
bin_name = ['Very low', 'Low', 'Average', 'High', 'Very high']
data['salary_group'] = pd.cut(data['salary'], bins, labels=bin_name, include_lowest=True)
data.head()
```
