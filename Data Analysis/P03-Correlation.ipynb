```python
import pandas as pd

data = pd.read_csv('correlation.csv')
data.head()
```

```python
data.describe()
```

```python
data.corr(method='pearson', min_periods=1)
```

```python
data.corr(method='kendall', min_periods=1)
```

```python
data.corr(method='spearman', min_periods=1)
```

### Positively Correlated

```python
data[['Temp', 'Cold_Drink']].corr()
```

### Negatively Correlated

```python
data[['Temp', 'Hot_Drink']].corr()
```

### No Correlated

```python
data[['Temp', 'Snacks']].corr()
```

```python
import matplotlib.pyplot as plt

data.corr().style.background_gradient(cmap='coolwarm')
```

```python
data.corr().style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
data.corr().style.background_gradient(cmap='coolwarm').format("{:.2f}")
```
