```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('regression-one.csv') 
df.head()
```

```python
df.rename(columns = {'Life expectancy': 'life-expect'}, inplace=True)
df.head()
```

```python
df = df.dropna()
df.shape
```

```python
df[df.isnull().any(axis=1)]
```

```python
X = df[['GDP', 'Alcohol', 'BMI', 'Schooling']]
Y = df['life-expect']
```

```python
from sklearn.linear_model import LinearRegression
ln = LinearRegression()
ln.fit(X,Y)
```

```python
ln.intercept_
```

```python
ln.coef_
```

```python
Yhat = ln.predict(X)
```

```python
data = pd.DataFrame({'Actual': df['life-expect'], 'Predicted': Yhat})
data.head(20)
```

### Distribution plot

```python
import seaborn as sns

sns.kdeplot(df['life-expect'], color='r', label='Actual Values')
sns.kdeplot(Yhat, color='b', label='Fitted Values')

plt.title('Actual vs Fitted values for Life-expect')
plt.xlabel('Age range')
plt.legend()
plt.show()
```
