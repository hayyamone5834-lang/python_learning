```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('regression-one.csv')
df.head()
```

```python
df.rename(columns = {'Life expectancy': 'life-expect'}, inplace=True)
df = df.dropna()
df.shape
```

```python
df[df.isnull().any(axis=1)]
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slrX = df[['Schooling']]
Y = df['life-expect']
slr.fit(slrX, Y)
```

```python
slrYhat = slr.predict(slrX)
```

```python
print(slrYhat[0:4])
```

```python
from sklearn.metrics import mean_squared_error

slr_MSE = mean_squared_error(df['life-expect'], slrYhat)
print('MSE for Simple Linear Regression: ', slr_MSE)
```

```python
slr_RSquared = slr.score(slrX, Y)
print('RSquared for Simple Linear Regression: ', slr_RSquared)
```

```python
mlr = LinearRegression()
mlrX = df[['GDP', 'Alcohol', 'BMI', 'Schooling']]
mlr.fit(mlrX, Y)
```

```python
mlrYhat = mlr.predict(mlrX)
print(mlrYhat[0:4])
```

```python
mlr_MSE = mean_squared_error(df['life-expect'], mlrYhat)
print('MSE for Multiple Linear Regression: ', mlr_MSE)
```

```python
mlr_RSquared = mlr.score(mlrX, Y)
print('RSquared for Multiple Linear Regression: ', mlr_RSquared)
```

```python
prX = df['Schooling']
f = np.polyfit(prX,Y,3)
pr = np.poly1d(f)
```

```python
prYhat = pr(prX)
print(prYhat[0:4])
```

```python
pr_MSE = mean_squared_error(df['life-expect'], prYhat)
print('MSE for Polynomial Regression: ', pr_MSE)
```

```python
from sklearn.metrics import r2_score

pr_RSquared = r2_score(df['life-expect'], prYhat)
print('RSquared for Polynomial Regression: ', pr_RSquared)
```
